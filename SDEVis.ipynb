```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the SDE
sigma = 25  # Standard deviation
T = 2.0  # Total time
dt = 0.01  # Time step
N = int(T / dt)  # Number of time steps
num_trajectories = 10  # Number of trajectories to simulate

# Time vector
time = np.linspace(0, T, N)

# Initialize an array to store the trajectories
trajectories = np.zeros((num_trajectories, N))

# Simulate the trajectories
for i in range(num_trajectories):
    # Initialize the current trajectory
    x_t = 0
    for j in range(1, N):
        # The increment of a Wiener process is normally distributed
        dw = np.random.normal(0, np.sqrt(dt))
        # Update the trajectory using the SDE
        x_t += sigma ** time[j] * dw
        trajectories[i, j] = x_t

# Plot the trajectories
plt.figure(figsize=(12, 6))
for i in range(num_trajectories):
    plt.plot(time, trajectories[i], label=f'Trajectory {i+1}')
plt.title('Trajectories of the SDE $dx = \sigma^t dW$')
plt.xlabel('Time t')
plt.ylabel('X(t)')
plt.legend()
plt.show()
```

```python
# Parameters for the new SDE
T = 2.0  # Total time
dt = 0.01  # Time step
N = int(T / dt)  # Number of time steps
num_trajectories = 10  # Number of trajectories to simulate

# Time vector
time = np.linspace(0, T, N)

# Initialize an array to store the trajectories
trajectories = np.zeros((num_trajectories, N))

# Simulate the trajectories
for i in range(num_trajectories):
    # Initialize the current trajectory
    x_t = 0
    for j in range(1, N):
        # The increment of a Wiener process is normally distributed
        dw = np.random.normal(0, np.sqrt(dt))
        # Update the trajectory using the SDE
        # dx = -1/2 * t * x * dt + sqrt(t) * dw
        x_t += -1/2 * time[j] * x_t * dt + np.sqrt(time[j]) * dw
        trajectories[i, j] = x_t

# Plot the trajectories
plt.figure(figsize=(12, 6))
for i in range(num_trajectories):
    plt.plot(time, trajectories[i], label=f'Trajectory {i+1}')
plt.title('Trajectories of the SDE $dx = -\\frac{1}{2} t x dt + \sqrt{t} dW$')
plt.xlabel('Time t')
plt.ylabel('X(t)')
plt.legend()
plt.show()
```
